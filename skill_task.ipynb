```python
train.replace([np.inf, -np.inf], np.nan, inplace=True)
test.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
columns = set(test.isna().sum()[test.isna().sum() == len(test)].index).union(
    set(train.isna().sum()[train.isna().sum() == len(train)].index))
```

```python
train.drop(columns=columns, inplace=True)
test.drop(columns=columns, inplace=True)
```

```python
# cols_train = set(train.isna().sum()[train.isna().sum() > 0.1*len(train)].index)
# cols_test = set(test.isna().sum()[test.isna().sum() > 0.1*len(test)].index)
```

```python
# train.drop(columns=cols_train, inplace=True)
# test.drop(columns=cols_train, inplace=True)
```

```python
columns_zero = list(set(test.isna().sum()[test.isna().sum() >= 0.3*len(test)].index).union(
    set(train.isna().sum()[train.isna().sum() >= 0.3*len(train)].index)))
```

```python
train[columns_zero] = train[columns_zero].fillna(0)
test[columns_zero] = test[columns_zero].fillna(0)
```

```python
train.fillna(train.median(), inplace=True)
test.fillna(test.median(), inplace=True)
```

```python
model = nn.Sequential(
    nn.Linear(1534, 3200, bias=True),
    nn.LeakyReLU(0.1),
    nn.BatchNorm1d(3200),
    nn.Dropout(0.5),
    nn.Linear(3200, 1, bias=True),
    nn.Sigmoid()
)
```

```python
def init_weights(m):
    if type(m) == nn.Linear:
        torch.nn.init.kaiming_uniform_(m.weight)
        m.bias.data.fill_(0.01)
```

```python
model.apply(init_weights)
```

```python
X = train.drop(columns='y')
y = train['y']
X_test = test
```

```python
ss = StandardScaler()
X = ss.fit_transform(X)
X_test = ss.transform(X_test)
```

```python
# pca = PCA(n_components = 100, random_state=7)
# X = pca.fit_transform(X)
# X_test = pca.transform(X_test)
```

```python
X.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=77)
```

```python
X_test.shape
```

```python
X_train_torch = torch.FloatTensor(X_train)
X_val_torch = torch.FloatTensor(X_val)
y_train = torch.FloatTensor(y_train)
y_val = torch.FloatTensor(y_val)
```

```python
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-5, weight_decay=1e-5)
# scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=20, verbose=True, factor=0.5)
# scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, 10)
```

```python
batch_size = 100
num_epochs = 200
early_stopping_rounds = 20
```

```python
X_train_torch
```

```python
model.train()
losses = []
losses_val = []
roc_auc_train = []
roc_auc_val = []
best_score = 0
best_model = None
counter = 0
for e in range(num_epochs):
    print('EPOCH: ', e)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, 5)
    for i in range(0, X_train_torch.shape[0], batch_size):
        scheduler.step()
        X_batch = X_train_torch[i:i+batch_size, :]
        y_batch = y_train[i:i+batch_size]
        optimizer.zero_grad()
        preds = model(X_batch)
        
        loss = criterion(preds, y_batch.view(preds.shape))
        loss.backward()
        optimizer.step()
#         losses.append(loss)
        
        
    pred_train = model(X_train_torch)
    loss_train = criterion(pred_train, y_train.view(pred_train.shape))
    losses.append(loss_train)
    print('LOSS TRAIN:', loss_train)
    score = roc_auc_score(y_train.numpy().astype(int), pred_train.detach().numpy().reshape(-1))
    print('ROC_AUC TRAIN:', score)
    roc_auc_train.append(score)
    
    pred_val = model(X_val_torch)
    loss_val = criterion(pred_val, y_val.view(pred_val.shape))
    losses_val.append(loss_val)
    print('LOSS VAL: ', loss_val)
    score = roc_auc_score(y_val.numpy().astype(int), pred_val.detach().numpy().reshape(-1))
    print('ROC_AUC VAL:', score)
    roc_auc_val.append(score)
    if score > best_score:
        print('ROC_AUC INCREASE: {}'.format(score - best_score))
        counter = 0
        best_score = score
        best_model = model
    else: 
        counter += 1
    if counter == early_stopping_rounds:
        break
    
    
```

```python
plt.figure(figsize=(10, 7))
plt.plot(roc_auc_train)
plt.plot(roc_auc_val)
plt.legend(['train', 'val'])
```

```python
plt.figure(figsize=(10, 7))
plt.plot(losses)
plt.plot(losses_val)
plt.legend(['train', 'val'])
```
